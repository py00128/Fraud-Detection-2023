```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import RobustScaler
from sklearn.model_selection import train_test_split
import seaborn as sns
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, roc_auc_score
```

```python
df = pd.read_csv("creditcard_2023.csv")
```

```python
missing_values = df.isna()

missing_count = missing_values.sum()

total_missing_count = missing_values.sum().sum()
```

```python
X = df.drop(["id", "Class"], axis = 1)
y = df["Class"]
```

```python
df.corr()
```

```python
scaler = RobustScaler()
X = scaler.fit_transform(X)

X_train, X_t, y_train, y_t = train_test_split(X, y, test_size = 0.3, shuffle= True)
X_valid, X_test, y_valid, y_test = train_test_split(X_t, y_t, test_size = 0.5, shuffle= True)
```

```python
print("train: ", X_train.shape)
print("valid: ", X_valid.shape)
print("test: ", X_test.shape)
```

```python
print("train: ", y_train.shape)
print("valid: ", y_valid.shape)
print("test: ", y_test.shape)
```

# Logistic Regression

```python
model = LogisticRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_valid)

accuracy = accuracy_score(y_valid, y_pred)
classification_rep = classification_report(y_valid, y_pred)
roc_auc = roc_auc_score(y_valid, y_pred)

print(f"Accuracy: {accuracy}")
print("Classification Report:")
print(classification_rep)
print(f"ROC AUC: {roc_auc}")
```

# Random Forest

```python
model = RandomForestClassifier()
model.fit(X_train, y_train)

y_pred = model.predict(X_valid)

accuracy = accuracy_score(y_valid, y_pred)
classification_rep = classification_report(y_valid, y_pred)
roc_auc = roc_auc_score(y_valid, y_pred)

print(f"Accuracy: {accuracy}")
print("Classification Report:")
print(classification_rep)
print(f"ROC AUC: {roc_auc}")
```

# Support Vector Machine (SVM)

```python
model = SVC()
model.fit(X_train, y_train)

y_pred = model.predict(X_valid)

accuracy = accuracy_score(y_valid, y_pred)
classification_rep = classification_report(y_valid, y_pred)
roc_auc = roc_auc_score(y_valid, y_pred)

print(f"Accuracy: {accuracy}")
print("Classification Report:")
print(classification_rep)
print(f"ROC AUC: {roc_auc}")
```

# Neural Network

```python
model = keras.Sequential([
    keras.layers.Input(shape=(X_train.shape[1],)), 
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dense(32, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

model.fit(X_train, y_train, epochs=10, batch_size=32, validation_data=(X_valid, y_valid))

loss, accuracy = model.evaluate(X_test, y_test)

print(f"Test loss: {loss:.4f}")
print(f"Test accuracy: {accuracy:.4f}")
```
